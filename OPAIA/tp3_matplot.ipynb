# Exercise 1: Simple Polynomial Plot



Plot a Polynomial Function:
- Create an array x with 100 linearly spaced points between -10 and 10.
- Create an array y for the polynomial function y = 2x^3 - 5x^2 + 3x - 7.
- Plot y versus x with a blue line.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = 2 * x**3 - 5 * x**2 + 3 * x - 7

plt.plot(x, y, color='blue')

plt.title("Polynomial Function: $y = 2x^3 - 5x^2 + 3x - 7$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.show()
```

# Exercise 2: Exponential and Logarithmic Plot
Plot Exponential and Logarithmic Functions:
- Create an array x with 500 linearly spaced points between 0.1 and 10.
- Compute y1 = exp(x) and y2 = log(x).
- Plot both functions on the same graph using different colors and line styles.

```python
x = np.linspace(0.1, 10, 500)

y1 = np.exp(x)
y2 = np.log(x)

plt.figure(figsize=(10, 6))

plt.plot(x, y1, color='blue', linestyle='-', label='y = exp(x)')
plt.plot(x, y2, color='orange', linestyle='--', label='y = log(x)')

plt.grid()
plt.title("Exponential and Logarithmic Functions")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)

plt.legend()
plt.savefig("exponential_logarithmic_functions.png")
plt.show()
```

# Exercise 3: Figure and Subplots
Create Subplots with Different Functions:
- Create a figure with two subplots in one row.
- In the first subplot, plot y = tan(x) and y = arctan(x) for x ranging from -2π to 2π.
- In the second subplot, plot y = sinh(x) and y = cosh(x) for x ranging from -2 to 2.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

x1 = np.linspace(-2 * np.pi, 2 * np.pi, 500)

ax1.plot(x1, np.tan(x1), label='y = tan(x)', color='blue')
ax1.plot(x1, np.arctan(x1), label='y = arctan(x)', color='orange')
ax1.set_title("Tangent and Arctangent Functions")
ax1.set_xlabel("x", fontsize=12)
ax1.set_ylabel("y", fontsize=12)
ax1.legend()
ax1.grid()
ax1.set_ylim(-10, 10)  

x2 = np.linspace(-2, 2, 500)

ax2.plot(x2, np.sinh(x2), label='y = sinh(x)', color='green')
ax2.plot(x2, np.cosh(x2), label='y = cosh(x)', color='red')
ax2.set_title("Hyperbolic Sine and Cosine Functions")
ax2.set_xlabel("x", fontsize=12)
ax2.set_ylabel("y", fontsize=12)
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()
```

# Exercise 5: Scatter Plot
Plot a Scatter Plot with Mathematical Functions:
- Create two arrays x and y, each with 500 values, where x is uniformly distributed and y = sin(x) + random noise.
- Plot a scatter plot of y versus x with varying marker sizes and colors based on the values of y.

```python
x = np.linspace(0, 10, 500)

y = np.sin(x) + np.random.normal(scale=0.1, size=x.shape)

sizes = 50 + 200 * (y - np.min(y)) / (np.max(y) - np.min(y))  
colors = y 

plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.6, edgecolors='w')

plt.xticks([])
plt.yticks([])
plt.title("Scatter Plot of $y = \sin(x) + \text{noise}$", fontsize=14)
plt.savefig("scatter_plot_sin_noise.pdf")

plt.show()
```

# Exercise 6: Contour Plot
Plot a Contour Plot of a Mathematical Function:
- Create an array x of 200 points linearly spaced between -5 and 5, and a similar array y.
- Create a 2D grid with numpy.meshgrid and define the function f(x, y) = sin(sqrt(x^2 + y^2)).
- Plot a contour plot of this function.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

plt.figure(figsize=(10, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='plasma')

plt.clabel(contour, inline=True, fontsize=8)
plt.title("Contour Plot of $f(x, y) = \sin(\sqrt{x^2 + y^2})$", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)

plt.grid()
plt.show()
```

# Exercise 7: 3D Surface Plot
Plot a 3D Surface Plot of a Mathematical Function:
- Create arrays X and Y with values ranging from -5 to 5 with a step of 0.25.
- Compute Z = cos(sqrt(X^2 + Y^2)).
- Plot a 3D surface plot of Z.

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-5, 5.25, 0.25)
y = np.arange(-5, 5.25, 0.25)
X, Y = np.meshgrid(x, y)
Z = np.cos(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.view_init(elev=30, azim=45)
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of Z = cos(sqrt(X^2 + Y^2))')
plt.show()
```

# Exercise 10: Changing Viewing Angle
Change the Viewing Angle of a 3D Plot:
- Reuse the 3D surface plot from Exercise 7.
- Create two subplots with different viewing angles using view_init.

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-5, 5.25, 0.25)
y = np.arange(-5, 5.25, 0.25)
X, Y = np.meshgrid(x, y)

Z = np.cos(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(12, 6))

ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax1.view_init(elev=30, azim=30)  
ax1.set_title('Viewing Angle: Elevation 30°, Azimuth 30°')
ax2 = fig.add_subplot(122, projection='3d')
surf2 = ax2.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax2.view_init(elev=60, azim=120)   
ax2.set_title('Viewing Angle: Elevation 60°, Azimuth 120°')
fig.colorbar(surf2, ax=ax2, shrink=0.5, aspect=10)
plt.tight_layout()
plt.show()
```

# Exercise 14: 3D Bar Plot
Plot a 3D Bar Plot:
- Create arrays x and y with 10 linearly spaced points between -5 and 5.
- Compute z = exp(-0.1 * (x^2 + y^2)).
- Plot a 3D bar plot of z values.

```python

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

X, Y = np.meshgrid(x, y)
Z = np.exp(-0.1 * (X**2 + Y**2))

fig = plt.figure(figsize=(10, 7))

ax = fig.add_subplot(111, projection='3d')

ax.bar3d(X.flatten(), Y.flatten(), np.zeros_like(Z.flatten()),
         dx=0.8, dy=0.8, dz=Z.flatten(),
         color='cyan', edgecolor='k')

mappable = plt.cm.ScalarMappable(cmap='viridis')
mappable.set_array(Z)
plt.colorbar(mappable, ax=ax, shrink=0.5, aspect=10, label='Z values')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Bar Plot of Z = exp(-0.1 * (X^2 + Y^2))')

# Change the viewing angle
ax.view_init(elev=30, azim=45)

plt.show()
```

# Exercise 15: 3D Vector Field
Plot a 3D Vector Field:
- Create arrays X, Y, and Z with values ranging from -5 to 5.

- Plot a 3D vector field using quiver3D.

```python
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
z = np.linspace(-5, 5, 10)

X, Y, Z = np.meshgrid(x, y, z)
U = -Y
V = X
W = Z
fig = plt.figure(figsize=(12, 8))

ax = fig.add_subplot(111, projection='3d')

scale = 5   
ax.quiver(X, Y, Z, U, V, W, length=0.5, color='blue', normalize=True)

ax.set_title('3D Vector Field: U = -Y, V = X, W = Z')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

# Exercise 17: 3D Line Plot
Plot a 3D Line Plot:
- Create an array t with 100 points linearly spaced between 0 and 4π.
- Compute X = sin(t), Y = cos(t), and Z = t.
- Plot a 3D line plot of X, Y, and Z.


```python
from scipy.stats import gaussian_kde

x = np.random.normal(loc=0, scale=1, size=1000)
y = np.random.normal(loc=0, scale=1, size=1000)
z = np.random.normal(loc=0, scale=1, size=1000)
xyz = np.vstack([x, y, z])
kde = gaussian_kde(xyz)

xi, yi, zi = np.mgrid[-3:3:30j, -3:3:30j, -3:3:30j]
density = kde(np.vstack([xi.ravel(), yi.ravel(), zi.ravel()]))
density = density.reshape(xi.shape)

fig = plt.figure(figsize=(10, 7))

ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o', alpha=0.1)  

norm_density = (density - np.min(density)) / (np.max(density) - np.min(density))
ax.scatter(xi, yi, zi, c=norm_density.ravel(), cmap='viridis', alpha=0.5)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Density Plot of Random Points')
plt.show()
```

# Exercise 20: 3D Density Plot
- Plot a 3D Density Plot:
- Generate 1000 random points in 3D space.
- Create a density plot to show where points are concentrated.

```python
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import gaussian_kde

x = np.random.uniform(-5, 5, 1000)
y = np.random.uniform(-5, 5, 1000)
z = np.random.uniform(-5, 5, 1000)

xyz = np.vstack([x, y, z])   
kde = gaussian_kde(xyz) 

density = kde(xyz)  

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(x, y, z, c=density, cmap='viridis', alpha=0.5)

cbar = plt.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Density')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Density Plot of Random Points')

plt.show()
```
