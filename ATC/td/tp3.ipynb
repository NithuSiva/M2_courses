```python
def load_data(dist_df,
             ang_df,
             lab_df,
             land_df,
             xyz_df):
    X = []
    y = []
    for idx, row in lab_df.iterrows():
        y.append(row.pose)
        tmp_X = [
                 dist_df[dist_df.pose_id==row.pose_id].iloc[:,1:].values,
                 ang_df[ang_df.pose_id==row.pose_id].iloc[:,1:].values,
                 land_df[land_df.pose_id==row.pose_id].iloc[:,1:].values,
                 xyz_df[xyz_df.pose_id==row.pose_id].iloc[:,1:].values
                ]
        X.append(tmp_X)
    return X, y
```

```python
X, y = load_data(d_distances_df,
                 angles_df,
                 labels_df,
                 landmarks_df,
                 xyz_distances_df)
```
